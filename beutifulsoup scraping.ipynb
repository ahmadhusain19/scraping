```python
from bs4 import BeautifulSoup
from requests import get
```

```python
# URL IMDb untuk film tahun 2017 dengan sorting berdasarkan jumlah vote terbanyak
url = "https://www.imdb.com/search/title/?genres=sci-fi"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"}
```

```python
response=get(url, headers=headers)
soup=BeautifulSoup(response.text, 'html.parser')
```

```python
cek = soup.find_all("div", class_="sc-d5ea4b9d-0 ejavrk")
```

```python
for x in cek:
    print(x.prettify())
```

```python
for x in cek:
  title_element = x.find("h3", class_="ipc-title__text")
  print(title_element.text)
```

```python
#Memanggil container dari deskripsi
deskripsi_containers=soup.find_all("div", class_="sc-d49a611d-1 csbzhP")
```

```python
for x in deskripsi_containers:
    print(x.prettify())
```

```python
for x in deskripsi_containers:
  tahun = x.find("div", class_="ipc-html-content-inner-div")
  print(tahun.text)
```

```python
# Menyimpan hasil scraping
film_data = []
for x in cek:
    title = x.find("h3", class_="ipc-title__text")
    tahun = x.find("span", class_="sc-d5ea4b9d-7 URyjV dli-title-metadata-item")
    raiting = x.find("span", class_="ipc-rating-star--rating")
    viewers = x.find("span", class_="ipc-rating-star--voteCount")

    film_data.append({
        "Title": title.text,
        "Year": tahun.text,
        "raiting": raiting.text,
        "viewers": viewers.text})
```

```python
print(film_data)
```

```python
narasi = []
for x in deskripsi_containers:
  deskripsi = x.find("div", class_="ipc-html-content-inner-div")

  narasi.append({
        "narasi": deskripsi.text})
```

Menyimpan data dalam format json

```python
import json

# Menyimpan hasil scraping
film_data = []
for x in cek:
    title = x.find("h3", class_="ipc-title__text")
    tahun = x.find("span", class_="sc-d5ea4b9d-7 URyjV dli-title-metadata-item")
    raiting = x.find("span", class_="ipc-rating-star--rating")
    viewers = x.find("span", class_="ipc-rating-star--voteCount")

    film_data.append({
        "Title": title.text if title else "N/A",
        "Year": tahun.text if tahun else "N/A",
        "Rating": raiting.text if raiting else "N/A",
        "Viewers": viewers.text if viewers else "N/A"
    })

# Menyimpan ke file JSON
with open("imdb_data.json", "w", encoding="utf-8") as json_file:
    json.dump(film_data, json_file, indent=4, ensure_ascii=False)

print("Data berhasil disimpan ke imdb_data.json")
```
